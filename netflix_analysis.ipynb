```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv("netflix_titles.csv")

# Basic info
print("Dataset shape:", df.shape)
print(df.head())

# 1 Movie vs TV Show count
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='Set2')
plt.title("Movies vs TV Shows on Netflix")
plt.show()

# 2 Top 10 Genres
top_genres = df['listed_in'].value_counts().head(10)
top_genres.plot(kind='barh', color='coral')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Count")
plt.show()

# 3 Most Frequent Directors
top_directors = df['director'].value_counts().head(10)
top_directors.plot(kind='bar', color='lightblue')
plt.title("Top 10 Directors on Netflix")
plt.xticks(rotation=45)
plt.show()

# 4 Trend by Release Year
year_data = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(10,5))
plt.plot(year_data.index, year_data.values, marker='o')
plt.title("Netflix Content Over the Years")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.grid(True)
plt.show()
```
